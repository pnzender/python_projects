# Variables =   A container for a value (string, integer, float, boolean) 
#               A variable behaves as if it was the value that it contains

```python
# Strings
first_name = "Bonnie"
food = "soup"

#Integer
age = 6

# Float
weight = 14.5

# Boolean
is_cat = False
is_dog = True

print(f"Hello, my name is {first_name} and I like to eat {food}.")
print(f"I am {age} years old and I weigh {weight} pounds.")
print(f"Am I a cat? {is_cat}. Am I a dog? {is_dog}.")
```

# Tycasting =  The process of converting a value from one data type to another
#              str(), int(), float(), bool()

```python
age = 10.0
print(type(age))

age = int(age)
print(type(age))
```

#input() =  function that allows user to enter data
#           returns the entered data as a string

```python
input_name = input("What is your name? ")
input_age = input("How old are you? ")
age = int(input_age)
print(f"Hello, {input_name}!")
print(f"You are {age} years old.")

```

# Exercise 01 - Rectangle Area Calc

```python
length = float(input("What is the length of the rectangle? "))
width = float(input("What is the width of the rectangle? "))

area = length * width
print(f"The area of the rectangle is {area}.")
```
